#### melt 사용하여 정리

```python
weather_22
```

```python
# melt 사용하여 정리
# 지면온도, 지중온도를 합쳐서 지표온도라고 일컫는다
# 지표온도와 관련된 열을 제외하고 모두 id_vars에 넣는다
# var_name은 지표온도
# value_name은 온도로 한다.

# 현지기압 해면기압도 id_vars에서 뺄까 싶다가 온도와는 상관이 없기 때문에 그대로 둠
# 여기서 질문이 생김 만약 variable을 두개로 하고 싶으면 melt에는 기능이 없나?
# 즉 여기서 만약 기압도 현지기압 해면기압 두개가 아니라 다른 기압들까지해서 카테고리가 많아지면 이것도 분류하면 좋을텐데
# 이 경우에는 어떻게 해야하는지 궁금해짐. 단순 melt를 두번 반복?
weather_22_spl = pd.melt(weather_22, id_vars=['지점', '지점명', '일시', '기온(°C)', '풍속(m/s)', '풍향(16방위)', '습도(%)', '증기압(hPa)',
       '이슬점온도(°C)', '현지기압(hPa)', '해면기압(hPa)', '중하층운량(10분위)', '시정(10m)'], var_name = '지표온도 (°C)', value_name='온도 (°C)')
weather_22_spl
```

```python
# 여기서 copy paste함
weather_22.columns
```

```python
# 문자를 split하려 했지만
# 지면온도는 앞에 숫자가 없다는 문제가 있다.
variable_spl = weather_22_spl['지표온도 (°C)'].str.split(' ')
variable_spl
```

```python
# 지면온도는 땅의 온도 이고 지중온도는 땅 속의 온도라는 뜻이다
# 따라서 지면온도 앞에 0cm 문자를 추가해주고 다시 split을 진행한다.
weather_22_spl['지표온도 (°C)'] = weather_22_spl['지표온도 (°C)'].apply(lambda x: '0cm ' + x if x=='지면온도(°C)' else x)
weather_22_spl
```

```python
# split 다시 진행
variable_spl = weather_22_spl['지표온도 (°C)'].str.split(' ')
variable_spl
```

```python
depth = variable_spl.str.get(0)
depth
```

```python
temptype = variable_spl.str.get(1)
temptype
```

```python
# depth를 추가해준다.
weather_22_spl['depth'] = depth
```

```python
weather_22_spl
```

```python
# tempature type을 추가해준다
weather_22_spl['temp_type']=temptype
weather_22_spl
```

```python
# 똑같이 21년도에도 적용
weather_21_spl = pd.melt(weather_21, id_vars=['지점', '지점명', '일시', '기온(°C)', '풍속(m/s)', '풍향(16방위)', '습도(%)', '증기압(hPa)',
       '이슬점온도(°C)', '현지기압(hPa)', '해면기압(hPa)', '중하층운량(10분위)', '시정(10m)'], var_name = '지표온도 (°C)', value_name='온도 (°C)')
weather_21_spl
```

```python
weather_21_spl['지표온도 (°C)'] = weather_21_spl['지표온도 (°C)'].apply(lambda x: '0cm ' + x if x=='지면온도(°C)' else x)
weather_21_spl
```

```python
variable_spl21 = weather_21_spl['지표온도 (°C)'].str.split(' ')
variable_spl21
```

```python
depth21 = variable_spl21.str.get(0)
depth21
```

```python
temptype21 = variable_spl21.str.get(1)
temptype21
```

```python
weather_21_spl['depth'] = depth21
weather_21_spl['temp_type'] = temptype21
weather_21_spl
```

```python
# 느낀점은 melt 쓸 일이 많이 없을 것 같다는 것
# 그리고 특히 문자열 split해서 새로운 열 만들어주는 건 더더 많이 쓸 일 없을 것 같았다.
# 솔직히 이 날씨 데이터를 다루는 것도 땅 속 얼마나 깊이 들어간 것인지를 알고싶은 depth와 temp_type열이 굳이 필요한가 싶었다.
```
